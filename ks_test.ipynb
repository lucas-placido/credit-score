## KS (Kolmogorov-Smirnov)

“What is the probability that these two sets of samples were drawn from the same probability distribution?”

$$
D = \sup_{x} \left| F_{1}(x) - F_{2}(x) \right|
$$

```python
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_distributions(y, z):
    fig, ax = plt.subplots(2, 1, figsize=(8, 10))

    # Plot histograms
    ax[0].hist(y, bins=50, density=True, alpha=0.5, label='y ~ N(0, 1)')
    ax[0].hist(z, bins=50, density=True, alpha=0.5, label='z ~ N(1.1, 0.9)')
    ax[0].set_title('Histogram of Distributions')
    ax[0].set_xlabel('Value')
    ax[0].set_ylabel('Density')
    ax[0].legend()

    # Plot CDFs
    sorted_y = np.sort(y)
    sorted_z = np.sort(z)
    n_y = sorted_y.size
    n_z = sorted_z.size
    ax[1].plot(sorted_y, np.arange(1, n_y + 1) / n_y, label='y ~ N(0, 1)')
    ax[1].plot(sorted_z, np.arange(1, n_z + 1) / n_z, label='z ~ N(1.1, 0.9)')
    ax[1].set_title('CDF of Distributions')
    ax[1].set_xlabel('Value')
    ax[1].set_ylabel('Cumulative Probability')
    ax[1].legend()

    plt.tight_layout()
    plt.show()
```

```python
np.random.seed(12345678)
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 100000)
z = np.random.normal(1.1, 0.9, 10000)


print(ks_2samp(x, y))
print(ks_2samp(x, z))
```

```python
# Plot distributions
plot_distributions(x, y)
plot_distributions(x, z)
```
